```python
datos_total = pd.read_excel(r"C:\Users\Asus\Downloads\Consolidado.xlsx")
```

```python
datos_total.head()
```

```python
tabla_contingencia = pd.crosstab(datos_total['FECHA INICIAL'], [datos_total['TIPO'], datos_total['SUBTIPO'], datos_total['EDAD'], datos_total['FEMENINO'], datos_total['RANGO DE EDAD']], margins=True)
```

```python
tabla_contingencia
```

```python
# Utilizamos la función chi2_contingency para calcular el índice de Chi-cuadrado y el p-valor
chi2, p, dof, expected = chi2_contingency(tabla_contingencia)
# Si el p-valor es menor que 0.05, podemos decir que existe una relación significativa entre las variables
if p < 0.05:
    print("Existe una relación significativa entre las variables (p = {})".format(p))
else:
    print("No existe una relación significativa entre las variables (p = {})".format(p))
```

```python
mca = prince.MCA(
        n_components=2,
        n_iter=3,
        copy=True,
        check_input=True,
        engine='auto',
        random_state=42)
mca = mca.fit(tabla_contingencia)
```

```python
coor = mca.plot_coordinates(
        X=tabla_contingencia,
        ax=None,
        figsize=(6, 6),
        show_row_points=True,
        row_points_size=10,
        show_row_labels=False,
        show_column_points=True,
        column_points_size=30,
        show_column_labels=False,
        legend_n_cols=1)
coor.get_figure()
```
